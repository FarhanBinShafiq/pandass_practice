# Merge

# it will join two or more  df based on the df's common value.
# if one value is not common in df's that value can't add by merge method
### it works as like set--insection method
### pd.merge(dfs,dfs,dfs,on='the column name which based it will merged",how='inner(default/outer/left/right),indicator=True)
############ How--------> 
outer---->means it will take all the value of between df's if there is no value exists.it will be none
left-----> it will take the left dfs all value and right's common value
right-------->it will take the right dfs all value and left's common value
###### indicator=True--------->means that from which df's the the value is come it will show

#### suffixes---> if there is common columns name in the dfs ,in mergeing we want to show the both we use suffixes=('_left','_right') 
********* if there is no value of the column it will be NaN**********************

### pd.Grouper is a class in the pandas library used for grouping time-series data. It's particularly useful when you want to group your data by some time frequency, such as by day, week, month, etc.

```python
temperature_d = pd.DataFrame({
    "city": ["mumbai","delhi","banglore",'c'],
    "temperature": [32,45,30,10],
    "speed": [32,45,30,12]
})
temperature_d
```

```python
temperature_f = pd.DataFrame({
    "city": ["mumbai","delhi","banglore",'o'],
     "temperature": [32,45,30,10],
    "speed": [32,45,30,12],
})
temperature_f
```

```python
pd.merge(temperature_f,temperature_d,on='city',suffixes=('_left_df','_right_df'))
```

```python
pd.merge(temperature_f,temperature_d,on='city',how='outer',indicator=True)
```

```python
pd.merge(temperature_f,temperature_d,on='city',how='left')
```

```python
pd.merge(temperature_f,temperature_d,on='city',how='right')
```

# Pivot 



The pivot operation on a DataFrame involves restructuring data to facilitate better analysis or visualization. It typically involves transforming values between rows and columns, often making analysis more convenient. In Python using pandas, you can use functions like .pivot() or pivot_table() to perform this operation. These functions help in reshaping the data for easier analysis and interpretation.

#####Pivot groper------> to see the full months avrg by it.
he code df.pivot_table(index=pd.Grouper(freq='M', key='date'), columns='city') is creating a pivot table from a DataFrame df.

It organizes the data by month (freq='M') using the dates from the 'date' column.
Each row in the pivot table represents a month.
The columns of the pivot table are the unique values found in the 'city' column of the DataFrame.
So, essentially, this code is reshaping the DataFrame to show the relationship between months and cities, 
possibly aggregating data in the process, if there are multiple entries for the same city within a month


```python
dsfs = pd.read_csv("weather3.csv")
dsfs
```

```python
dsfs['date'] = pd.to_datetime(dsfs['date'])
```

```python

dsfs.pivot_table(index=pd.Grouper(freq='M',key='date'),columns='city')
#he code df.pivot_table(index=pd.Grouper(freq='M', key='date'), columns='city') is creating a pivot table from a DataFrame df.

#It organizes the data by month (freq='M') using the dates from the 'date' column.
#Each row in the pivot table represents a month.
#The columns of the pivot table are the unique values found in the 'city' column of the DataFrame.
#So, essentially, this code is reshaping the DataFrame to show the relationship between months and cities, possibly 
#aggregating data in the process, if there are multiple entries for the same city within a month
```
