### Make a column "Strawberry" which will contain dichotomous variable[1]. It will return 1 for observations that contain "Strawberries" and 0 otherwise.

```python
def dichotomous_variable(product_name):
    if product_name == 'Strawberries':
        return 1
    else:
        return 0
```

```python
def dichotomous_variable_multiple(product_name, fruit_name):
    if product_name == fruit_name:
        return 1
    else:
        return 0
```

```python
dichotomous_variable_multiple("Strawberries", "Banana")
```

```python
df
```

```python
df["Strawberries"] = df["productname"].apply(dichotomous_variable)
df
```

```python
unique_product_name = df["productname"].unique().tolist()
unique_product_name
```

```python
def dichotomous_variable_multiple(product_name, fruit_name):
    if product_name == fruit_name:
        return 1
    else:
        return 0 
```

```python
for product_name in unique_product_name:
    df[product_name] = df["productname"].apply(dichotomous_variable_multiple, args=(product_name,))
```
