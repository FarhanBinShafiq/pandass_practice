### Make a column "Strawberry" which will contain dichotomous variable[1]. It will return 1 for observations that contain "Strawberries" and 0 otherwise.

Solve the problem with fucntion.You can also solve the above problem without function if you know more better method(AS I don't know)!

```python
df["productname"].unique()
```

```python
def dichotomous_variable(product_name):
    if product_name == "Strawberries":
        return 1
    else:
        return 0
```
