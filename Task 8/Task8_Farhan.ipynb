# Make another column "Margin type" whiere value is "run" for the rows that are measured by "run" and value is "wicket" for the rows that are measure by "wickets" from the column Margin. 

```python
def m_type(margin):
    if "run" in margin:
        return "run"
    elif "wicket" in margin:
        return "wicket"
    else:
        return None

 
df['Margin type'] = df['Margin'].apply(m_type)

 
```

```python
df
```

# Make two dataframes from the run and wickets row type and make another column where Margin column is of integer type.

```python
def run(margin):
   
        if "run" in margin:
            return int(margin.split()[0])
    
        else:
             return 0

 
df['run'] = df['Margin'].apply(run)
```

```python
def wicket(margin):
    if "wicket" in margin:
        return int(margin.split()[0])
    
    else:
        return 0

 
df['wicket'] = df['Margin'].apply(wicket)
```

```python
df
```

 # (group_by) sum of total runs and wickets
 

```python
run= df['run'].sum()
print('Total Run :',run)    
```

```python
wicket= df['wicket'].sum()
print('Total wicket :',wicket)   
```
